## When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

## it loses bikes, i know when comparing the current bikes with installed bikes for the modified station

```python
station_df = station_df.assign(lost = lambda x: x["current_dockcount"] - x['install_dockcount'])
lost = station_df[station_df['lost'] != 0][['name',"install_dockcount",'current_dockcount','lost']]
avg_lost = lost["lost"].mean()
avg_lost
```
